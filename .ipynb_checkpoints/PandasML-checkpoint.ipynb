## Sorting/ Describing data

```python
df.sort_values(['Type 1','HP'],ascending=False) # Value of both are False
```

```python
df.sort_values(['Type 1','HP'],ascending=[0,1]) # It means the value of Type 1 is false and the value of HP is True 
```

## Making changes to the data

```python
#Adding a new column
df['Total']= df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Sp. Def']
df
```

```python
#Deleting a column
df = df.drop(['Total'], axis=1) # Axis 1 is columns ---> and axis 0 is rows â†“
df
```
